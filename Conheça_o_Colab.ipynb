# Conheça o Colab

```
import csv

dados = [
    ['Date', 'Calories'],
    ['1900/02/01', 19  ],
    ['1900/02/02', 20],
    ['1900/02/03', 13  ],
    ['1900/02/04', 58  ],
    ['1900/02/05', 56  ],
    ['1900/02/06', 23  ],
    ['1900/02/07', 22  ],
    ['1900/02/08', 42  ],
    ['1900/02/09', 42  ],
    ['1900/02/10', 26  ],
    ['1900/02/11', 25  ],
    ['1900/02/12', 17  ]
]

with open('dados.csv', mode='w', newline='', encoding='utf-8') as arquivo:
    escritor = csv.writer(arquivo)
    escritor.writerows(dados)
```

```
import pandas as pd

caminho_arquivo = 'dados.csv'

df = pd.read_csv(
    caminho_arquivo,
    sep=',',
    engine='python',
    encoding='utf-8'
)

print(df.head())

df.to_csv(
    caminho_arquivo,
    sep=';',
    index=False,
    encoding='utf-8'
)

print("Dados atualizados e salvos com sucesso.")

df_subconjunto = df[['Date', 'Calories']]

print(df_subconjunto.head())
print(f"Colunas selecionadas: {df_subconjunto.columns.tolist()}")

df_subconjunto.to_csv(
    'subconjunto_dados.csv',
    sep=';',
    index=False,
    encoding='utf-8'
)

print("Arquivo salvo com sucesso.")

pd.set_option('display.max_rows', 9999)

df.to_csv(
    'seu_arquivo.csv',
    sep=';',
    index=False,
    encoding='utf-8'
)

print(df.to_string())

print(df.head(10))

print(df.tail(10))

df.info()

total_linhas = df.shape[0]

total_colunas = df.shape[1]

dados_nulos = df.isnull().sum()
total_nulos_geral = dados_nulos.sum()

tipos_dados = df.dtypes

memoria_utilizada = df.memory_usage(deep=True).sum()

print(f"\n--- Resumo Extraído ---")
print(f"Total de linhas: {total_linhas}")
print(f"Total de colunas: {total_colunas}")
print(f"Total de dados nulos no conjunto: {total_nulos_geral}")
print(f"\nDetalhe de nulos por coluna:\n{dados_nulos}")
print(f"\nTipos de dados por coluna:\n{tipos_dados}")
print(f"\nMemória utilizada: {memoria_utilizada} bytes ({memoria_utilizada / 1024:.2f} KB)")
```
